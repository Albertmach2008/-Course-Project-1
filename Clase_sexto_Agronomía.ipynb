```python
import pandas as pd
from google.colab import data_table
import matplotlib.pyplot as plt
import seaborn as sns
data_table.enable_dataframe_formatter()

ruta1 = "/content/agricultura.csv"
agri = pd.read_csv(ruta1)

pd.DataFrame(agri)

#agri.head()
#agri.info()
#agri
#agri.describe()
#print(agri)
#display
```

```python
#Limpieza de datos

#Cambiar nombre de columnA
agri2=agri.rename({"ID Campa�a":"ID", "Producci�n (Tn)": "Produccion (Ha)"}, axis="columns")

agri.drop(agri[agri['Provincia'] == 'MISIONES'].index, inplace = True)
agri

#Operación con columnas
agri2["Dif.Sem_Cos_(Ha)"] = agri2["Sup. Sembrada (Ha)"]-agri2["Sup. Cosechada (Ha)"]
agri2["Dif.Sem_Cos_(Ha)"].info()
agri2

#Eliminar una columna
del agri2["Campana"]
agri2

agri2.drop(["ID"], axis=1, inplace=True)
agri2.shape

#Eliminar filas
agri3=agri2.drop([1,2], axis=0, inplace=False)
agri3
agri3.shape

#Transformar en Numérico
agri3["Produccion (Ha)"] = pd.to_numeric(agri3["Produccion (Ha)"], errors= "coerce")
agri3.info()
```

```python
#Revisar NA
#agri3.isnull().sum()

#Reemplazo NA
agri_na1=agri3.dropna(inplace=False)
agri3.dropna(subset=['Produccion (Ha)'], inplace=True)
agri3.isnull().sum()
```

```python
# Histogramas
plt.figure(figsize=(10, 6))

#Renombro una función
a = plt.hist

a(agri3["Temperatura (Ce)"], color="skyblue" ,edgecolor= "black")
plt.xlabel("Temperatura (ºC)")
plt.ylabel("Frecuencia")
plt.title("Temperatura registrada en las parcelas cultivadas")

#Guardar una imagen
#plt.savefig("my_plot.png")
```

```python
m1 = agri3['Sup. Sembrada (Ha)'].mean()
sd1 = agri3['Sup. Sembrada (Ha)'].std()
m2 = agri3['Sup. Cosechada (Ha)'].mean()
sd2 = agri3['Sup. Cosechada (Ha)'].std()
m3 = agri3['Produccion (Ha)'].mean()
sd3 = agri3['Produccion (Ha)'].std()
m4 = agri3['Rendimiento (Kg/Ha)'].mean()
sd4 = agri3['Rendimiento (Kg/Ha)'].std()

print(m1, sd1, m2, sd2, m3, sd3, m4, sd4)
```

```python
#TRANSFORMACIÓN

import numpy as np

agri3["log.Sup.Sembrada"]=np.log(1+agri3['Sup. Sembrada (Ha)'])
agri3["log.Sup.Cosechada"]=np.log(1+agri3['Sup. Cosechada (Ha)'])
agri3["log.Produccion"]=np.log(1+agri3["Produccion (Ha)"])
agri3["log.Rendimiento"]=np.sqrt(agri3['Rendimiento (Kg/Ha)'])
agri3

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

axs[0].hist(agri3['log.Sup.Sembrada'],  color='Yellow', edgecolor='black')
axs[0].set_title('Superficie sembrada')

axs[1].hist(agri3['log.Sup.Cosechada'], color='Pink', edgecolor='black')
axs[1].set_title("Superficie cosechada")

axs[2].hist(agri3["log.Produccion"], color='Red', edgecolor='black')
axs[2].set_title('Producción de cultivos')

axs[3].hist(agri3['log.Rendimiento'], color='Blue', edgecolor='black')
axs[3].set_title('Rendimiento de cultivos')

for ax in axs:
    ax.set_xlabel('Valores')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()

plt.show()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(10, 5))

axs[0].boxplot(agri3["log.Sup.Sembrada"])
axs[0].set_title('Superficie sembrada')

axs[1].boxplot(agri3["log.Sup.Cosechada"])
axs[1].set_title("Superficie cosechada")

axs[2].boxplot(agri3["log.Produccion"])
axs[2].set_title('Producción de cultivos')

axs[3].boxplot(agri3['log.Rendimiento'])
axs[3].set_title('Rendimiento de cultivos')

plt.show()
```

```python
#Calculo quartiles
q1=agri3['Rendimiento (Kg/Ha)'].quantile(0.25)
q3=agri3['Rendimiento (Kg/Ha)'].quantile(0.75)

#Calculo IQ
iqr=q3-q1

agri3['Rendimiento.iqr'] = agri3['Rendimiento (Kg/Ha)'].mask(agri3['Rendimiento (Kg/Ha)'] > 10, iqr, inplace=False)
```

```python
#Gráfica boxplot de base limpiada

fig, axs = plt.subplots(1, 4, figsize=(10, 5))

axs[0].boxplot(agri3["log.Sup.Sembrada"])
axs[0].set_title('Superficie sembrada')

axs[1].boxplot(agri3["log.Sup.Cosechada"])
axs[1].set_title("Superficie cosechada")

axs[2].boxplot(agri3["log.Produccion"])
axs[2].set_title('Producción de cultivos')

axs[3].boxplot(agri3['Rendimiento.iqr'])
axs[3].set_title('Rendimiento de cultivos')

plt.show()
```

```python
import seaborn as sns
mean_agri3=agri3.groupby("Provincia").mean(numeric_only=True)

sns.regplot(x='log.Sup.Sembrada', y='Rendimiento.iqr',fit_reg=True, data=mean_agri3)
plt.title('Relación Rendimiento con Superficie sembrada')
plt.xlabel("log(Superficie sembrada (Ha))")
plt.ylabel("Rendimiento (Tn)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import google.colab.data_table as dt

#HISTOGRAMAS

sp_c.describe()

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

axs[0].hist(sp_c['height'],  color='Yellow', edgecolor='black')
axs[0].set_title('Altura de individuo')

axs[1].hist(sp_c['crown_width'], color='Pink', edgecolor='black')
axs[1].set_title('Tamaño de la copa')

axs[2].hist(sp_c['dbh'], color='Red', edgecolor='black')
axs[2].set_title('DAP')

axs[3].hist(sp_c['basal_area'], color='Blue', edgecolor='black')
axs[3].set_title('Área basal')

for ax in axs:
    ax.set_xlabel('Valores')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()

plt.show()
```

```python
import numpy as np

#TRANSFORMACIÓN

sp_c["log.height"]=np.log10(sp_c["height"])
sp_c["log.crown"]=np.log10(sp_c["crown_width"])
sp_c["log.dbh"]=np.log10(sp_c["dbh"])
sp_c["log.basal"]=np.log10(sp_c["basal_area"])
sp_c.describe()

fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

axs[0].hist(sp_c['log.height'],  color='Yellow', edgecolor='black')
axs[0].set_title('Altura de individuo')

axs[1].hist(sp_c['log.crown'], color='Pink', edgecolor='black')
axs[1].set_title('Tamaño de la copa')

axs[2].hist(sp_c['log.dbh'], color='Red', edgecolor='black')
axs[2].set_title('DAP')

axs[3].hist(sp_c['log.basal'], color='Blue', edgecolor='black')
axs[3].set_title('Área basal')

for ax in axs:
    ax.set_xlabel('Valores')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()

plt.show()
```
